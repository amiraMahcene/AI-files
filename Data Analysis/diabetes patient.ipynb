```python
data = {
    'patient_id': [1, 2, 3, 4, 5, 5, 6, 7, 8, 9],
    'age': [45, 200, 30, 35, 50, 50, 28, 67, 80, None],
    'diabetes_type': ['Type2', 'Type1', 'type2', 'Type 1', 'Type2', 'Type2', 'Type3', 'Type1', None, 'Type2'],
    'blood_pressure': ['120/80', '140/90', '90/60', '13.3/8.7 kPa', '130/85', '130/85', '150/95', '160/100', '110/70', None],
    'glucose': [150, 300, 88, 200, None, 200, 250, 400, 180, 130],
    'last_visit_date': ['2023-05-12', '05/12/2023', '2023.06.01', '2023-07-20', '07/20/2023', '07/20/2023', '2023-08-15', None, '2023-09-01', '2023-10-10']
}
df = pd.DataFrame(data)
print("Raw Data:")
print(df)
```

```python
print("\n--- Remove Duplicates ---")
df = df.drop_duplicates().reset_index(drop=True)
print(f"Rows after removing duplicates: {len(df)}")
```

```python
df.loc[df['glucose'].isna(), 'glucose'] = 200.0
df
```

```python
print("\n--- Remove Duplicates ---")
df = df.drop_duplicates().reset_index(drop=True)
print(f"Rows after removing duplicates: {len(df)}")
df
```

```python
print("\n--- Clean Diabetes Type ---")
# Replace variations with standardized values
df['diabetes_type'] = df['diabetes_type'].str.replace(' ', '').str.lower()
df['diabetes_type'] = df['diabetes_type'].replace({
    'type1': 'Type1', 
    'type2': 'Type2',
    'type3': 'Other'  # Assume "Type3" is invalid; map to "Other"
})
print(df['diabetes_type'].value_counts())
df
```

```python
print("\n--- Handle Missing Values ---")
# Impute age with median
median_age = df['age'].median()
df['age'] = df['age'].fillna(median_age)

# Impute glucose with median
median_glucose = df['glucose'].median()
df['glucose'] = df['glucose'].fillna(median_glucose)

# Drop rows with missing blood_pressure (only 1 row here)
df = df.dropna(subset=['blood_pressure'])
df
```

```python
def clean_blood_pressure(value):
    value = str(value).strip().lower()  # Convert to lowercase and remove spaces
    
    if 'kpa' in value:
        value = value.replace(' kpa', '')  # Remove ' kPa' (handles lowercase too)
        systolic, diastolic = map(float, value.split('/'))  # Convert to float
        return f"{systolic * 7.5:.1f}/{diastolic * 7.5:.1f}"
    return value

# Apply cleaning function
df['blood_pressure'] = df['blood_pressure'].apply(clean_blood_pressure)

# Split into systolic and diastolic columns
df[['systolic', 'diastolic']] = df['blood_pressure'].str.split('/', expand=True).astype(float)

# Drop the original blood_pressure column
df = df.drop('blood_pressure', axis=1)

print(df)
```

```python
print("\n--- Remove Age Outliers ---")
df = df[df['age'] <= 120]
```

```python
df
```

```python
print("\n--- Clean Dates ---")
df['last_visit_date'] = pd.to_datetime(
    df['last_visit_date'], 
    errors='coerce',  # Convert invalid dates to NaT
    format='mixed'
)
# Drop rows with invalid/missing dates
df = df.dropna(subset=['last_visit_date'])
```

```python
df
```

```python
print("\n--- Final Validation ---")
print(df.info())
print("\nSummary Stats:")
print(df.describe())
print("\nDiabetes Type Counts:")
print(df['diabetes_type'].value_counts())
```
